```python
import requests
import time # Import the time module
currency_records= []

for i in range(1,6):
  url = (f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&per_page=250&order=market_cap_desc&page={i}&sparkline=False")
  response = requests.get(url)
  time.sleep(8)
  if response.status_code == 200:
    data = response.json()
    currency_records.extend(data)
    print(f"Page {i}: Fetched {len(data)} records. Total records so far: {len(currency_records)}")
  else:
    print(f"Page {i}: Failed to fetch data. Status code: {response.status_code}")
   # Add a 2-second delay between requests
print(f"\nFinal total records: {len(currency_records)}")
print(currency_records)
```

```python
%%writefile CrossMarketAnalysis.py

import streamlit as st
import pandas as pd
from datetime import date, timedelta

from streamlit_option_menu import option_menu
import sqlite3

# Must be the first Streamlit command used
st.set_page_config(layout="wide")

connection = sqlite3.connect("/content/drive/MyDrive/cross market database/crossmarketanalysis.db")
cursor = connection.cursor()

with st.sidebar:
    selected = option_menu("Navigation", ["Market overview","Queries","Top 3 crypto Analysis"],
        icons=['house', 'search','radio',], menu_icon="cast", default_index=1)

# Define default dates outside the conditional block
#today = date.today()
#tomorrow = today + timedelta(days=1)
today = date(2025, 3,1)
tomorrow = date(2025,3 ,31)
# Set more appropriate default dates for 'Top 3 crypto Analysis'
default_start_crypto = date(2025, 3, 1)
default_end_crypto = date(2025, 3, 31)

start_default = today
end_default = tomorrow

if selected == "Market overview":
    st.title("CROSS-Market OVER VIEW")
    st.write("Crypto+oil+stock Market |SQL Powered analysis")

    # Create two columns
    col1, col2 = st.columns(2)

    # Place a date input in the first column
    with col1:
        start_date = st.date_input("Start Date", value=start_default)

    # Place a date input in the second column
    with col2:
        end_date = st.date_input("End Date", value=end_default)

    # Optional: Add logic to ensure end_date is after start_date
    if start_date > end_date:
        st.error("Error: End date must be after start date.")
    else:
        query = f"""
            SELECT
                AVG(cd.price_usd) AS avg_bitcoin_price,
                AVG(od.price_usd) AS avg_oil_price,
                AVG(CASE WHEN s.ticker = '^GSPC' THEN s.close ELSE NULL END) AS avg_gspc_close,
                AVG(CASE WHEN s.ticker = '^NSEI' THEN s.close ELSE NULL END) AS avg_nsei_close
            FROM
                coins_data AS cd
            INNER JOIN
                oil_datas AS od ON cd.date = od.date
            INNER JOIN
                stocks AS s ON cd.date = s.date
            WHERE
                cd.coin_id = 'bitcoin'
                AND cd.date BETWEEN '{start_date}' AND '{end_date}'
                AND s.ticker IN ('^GSPC', '^NSEI');
            """
        try:
            df_averages = pd.read_sql(query, con=connection)

            if not df_averages.empty and df_averages.iloc[0]['avg_bitcoin_price'] is not None:
                avg_bitcoin_price = df_averages.iloc[0]['avg_bitcoin_price']
                avg_oil_price = df_averages.iloc[0]['avg_oil_price']
                avg_gspc_close = df_averages.iloc[0]['avg_gspc_close']
                avg_nsei_close = df_averages.iloc[0]['avg_nsei_close']

                col1, col2, col3, col4 = st.columns(4)
                with col1:
                    st.metric("Avg Bitcoin Price", f"{avg_bitcoin_price:,.2f}")
                with col2:
                    st.metric("Avg Oil Price", f"{avg_oil_price:,.2f}")
                with col3:
                    st.metric("Avg S&P 500 Close", f"{avg_gspc_close:,.2f}")
                with col4:
                    st.metric("Avg NSEI Close", f"{avg_nsei_close:,.2f}")
            else:
                st.warning("No data available for the selected date range for market overview.")
        except Exception as e:
            st.error(f"Error fetching data: {e}")

    st.divider()

    st.write("Daily Market Snapshot")
    Market_snapshot= pd.read_sql("""SELECT
                cd.date,
                cd.price_usd AS Bitcoin_price,
                od.price_usd AS Oil_price,
                MAX(CASE WHEN s.ticker = '^GSPC' THEN s.close ELSE NULL END) AS sp500,
                MAX(CASE WHEN s.ticker = '^NSEI' THEN s.close ELSE NULL END) AS nifty
            FROM
                coins_data AS cd
            INNER JOIN
                oil_datas AS od ON cd.date = od.date
            LEFT JOIN -- Use LEFT JOIN to include dates even if no stock data is available
                stocks AS s ON cd.date = s.date
            WHERE
                cd.coin_id = 'bitcoin'
            GROUP BY
                cd.date, cd.price_usd, od.price_usd
            ORDER BY
                cd.date;
            """, con=connection)
    st.dataframe(Market_snapshot)

if selected == "Queries":
     option = st.selectbox(
    "Select a query",
    ("Find the top 3 crypto curencies by market cap",
     "List all coins where circulating supply exceeds 90% of total supply",
     "Get coins that are within 10% of their all-time-high (ATH)",
     "Find the average market cap rank of coins with volume above $1B",
     "Get the most recently updated coin",
     "Find the highest daily price of Bitcoin in the last 365 days",
     "Calculate the average daily price of Ethereum in the past 1 year",
     "Show the daily price trend of Bitcoin in January 2025",
     "Find the coin with the highest average price over 1 year",
     "Get the % change in Bitcoin’s price between Sep 2024 and Sep 2025",
     "Find the highest oil price in the last 5 years",
     "Get the average oil price per year",
     "Show oil prices during COVID crash (March–April 2020)",
     "Find the lowest price of oil in the last 10 years",
     "Calculate the volatility of oil prices (max-min difference per year)",
     "Get all stock prices for a given ticker",
     "Find the highest closing price for NASDAQ (^IXIC)",
     "List top 5 days with highest price difference (high - low) for S&P 500 (^GSPC)",
     "Get monthly average closing price for each ticker",
     "Get average trading volume of NSEI in 2024",
     "Compare Bitcoin vs Oil average price in 2025",
     "Check if Bitcoin moves with S&P 500 (correlation idea)",
     "Correlate Bitcoin closing price with crude oil closing price (same date)",
     "Compare Ethereum and NASDAQ daily prices for 2025",
     "Compare stock prices (^GSPC) with crude oil prices on the same dates",
     "Compare top 3 coins daily price trend vs Nifty (^NSEI)",
     "Top 3 crypto coins with stock indices for 2025",
     "Compare NASDAQ (^IXIC) with Ethereum price trends",
     "stock prices, oil prices, and Bitcoin prices for daily comparison",
     "Display price"),
)
     if option == "Find the top 3 crypto curencies by market cap":
         df = pd.read_sql("SELECT * FROM currency_data WHERE market_cap_rank <=3", con=connection)
         st.dataframe(df)
     if option ==  "List all coins where circulating supply exceeds 90% of total supply":
         df = pd.read_sql("SELECT * FROM currency_data WHERE circulating_supply >= (total_supply*0.9)", con=connection)
         st.dataframe(df)
     if option ==  "Get coins that are within 10% of their all-time-high (ATH)":
         df = pd.read_sql("SELECT * FROM currency_data WHERE circulating_supply <= (ath*0.1)", con=connection)
         st.dataframe(df)
     if option ==  "Find the average market cap rank of coins with volume above $1B":
         df = pd.read_sql("SELECT AVG(market_cap_rank) AS Average_Market_rank_coins FROM currency_data WHERE total_volume > 1000000000", con=connection)
         st.dataframe(df)
     if option ==  "Get the most recently updated coin":
         df = pd.read_sql("SELECT * FROM currency_data ORDER bY date DESC LIMIT 1", con=connection)
         row = cursor.fetchall()
         st.dataframe(df)
     if option =="Find the highest daily price of Bitcoin in the last 365 days":
         df = pd.read_sql("SELECT MAX(price_usd) AS Maximum_price From coins_data WHERE coin_id = 'bitcoin' ",con=connection)
         st.dataframe(df)
     if option =="Calculate the average daily price of Ethereum in the past 1 year":
         df = pd.read_sql("SELECT AVG(price_usd) AS Average_price_usd From coins_data WHERE coin_id = 'ethereum'",con=connection)
         st.dataframe(df)
     if option =="Show the daily price trend of Bitcoin in January 2025":
         df = pd.read_sql("SELECT date,price_usd FROM coins_data WHERE coin_id ='bitcoin' AND date BETWEEN '2025-01-01' AND '2025-12-31'",con=connection)
         st.dataframe(df)
     if option =="Find the coin with the highest average price over 1 year":
         df = pd.read_sql("SELECT coin_id,avg(price_usd) AS Average_coins_price_usd From coins_data GROUP BY coin_id ORDER BY AVG(price_usd) DESC LIMIT 1 ",con=connection)
         st.dataframe(df)
     if option =="Get the % change in Bitcoin’s price between Sep 2024 and Sep 2025":
         df = pd.read_sql("SELECT (MAX(price_usd)- MIN(price_usd))/Min(price_usd)*100 As PCT_change From coins_data WHERE coin_id = 'bitcoin' ",con=connection)
         st.dataframe(df)
     if option =="Find the highest oil price in the last 5 years":
         df = pd.read_sql("SELECT max(price_usd) from oil_datas WHERE date>= DATE('now', '-5 years')",con=connection)
         st.dataframe(df)
     if option =="Get the average oil price per year":
         df = pd.read_sql("SELECT avg(price_usd) FROM oil_datas WHERE date BETWEEN '2025-01-01' AND '2026-01-02'",con=connection)
         df1= pd.read_sql("SELECT strftime('%Y',DATE(date)) AS year, AVG(price_usd) AS avg_price FROM oil_datas GROUP BY year", con=connection)
         st.dataframe(df)
         st.dataframe(df1)
     if option =="Show oil prices during COVID crash (March–April 2020)":
         df = pd.read_sql("SELECT date, price_usd from oil_datas WHERE date BETWEEN '2020-03-01' AND '2020-04-01' ",con=connection)
         st.dataframe(df)
     if option =="Find the lowest price of oil in the last 10 years":
         df = pd.read_sql("SELECT min(price_usd) FROM oil_datas WHERE date>= DATE('now', '-10 years') ",con=connection)
         st.dataframe(df)
     if option =="Calculate the volatility of oil prices (max-min difference per year)":
         df = pd.read_sql("SELECT max(price_usd)-min(price_usd)  AS Volatility_Of_Price FROM oil_datas WHERE date BETWEEN '2025-01-01' AND '2026-01-02' ",con=connection)
         st.dataframe(df)
     if option =="Get all stock prices for a given ticker":
         df = pd.read_sql("SELECT open,high,low,close from stocks where ticker = '{i}' ",con=connection)
         st.dataframe(df)
     if option =="Find the highest closing price for NASDAQ (^IXIC)":
         df = pd.read_sql("SELECT MAX(close) From stocks WHERE ticker = 'IXIC' ",con=connection)
         st.dataframe(df)
     if option =="List top 5 days with highest price difference (high - low) for S&P 500 (^GSPC)":
         df = pd.read_sql("SELECT date,(high -low) as pricedifference from stocks where ticker = 'GSPC' ORDER BY pricedifference DESC LIMIT 5",con=connection)
         st.dataframe(df)
     if option =="Get monthly average closing price for each ticker":
         df = pd.read_sql("SELECT ticker,strftime('%Y-%m',date) As month,AVG(close) from stocks GROUP BY ticker, month ORDER BY ticker,month",con=connection)
         st.dataframe(df)
     if option =="Get average trading volume of NSEI in 2024":
         df = pd.read_sql("SELECT AVG(volume) FROM stocks WHERE ticker = '^NSEI' AND strftime('%Y', date) = '2024' ",con=connection)
         st.dataframe(df)
     if option =="Compare Bitcoin vs Oil average price in 2025":
         df = pd.read_sql("SELECT AVG(oil_datas.price_usd) AS Average_price_oil_USD, AVG(coins_data.price_usd) AS Average_Price_BTC_USd FROM oil_datas INNER JOIN coins_data ON coins_data.date = oil_datas.date WHERE coins_data.coin_id='bitcoin' ",con=connection)
         st.dataframe(df)
     if option =="Check if Bitcoin moves with S&P 500 (correlation idea)":
         df = pd.read_sql("SELECT coins_data.price_usd AS BTC_USD, stocks.close AS close_price FROM coins_data INNER JOIN stocks ON coins_data.date = stocks.date WHERE coins_data.coin_id = 'bitcoin' AND stocks.ticker = '^GSPC'",con=connection)
         st.dataframe(df)
         corr = df['BTC_USD'].corr(df['close_price'])
         st.metric(label = "correlation for year ", value = f'{corr:,.2f}')
     if option =="Compare Ethereum and NASDAQ daily prices for 2025":
         df = pd.read_sql( """SELECT cd.date,
                   cd.price_usd AS coins_price,
                   s.close AS IXIC_prie,
                   ((cd.price_usd - LAG(cd.price_usd, 1) OVER (PARTITION BY cd.coin_id ORDER BY cd.date)) * 100.0) / LAG(cd.price_usd, 1) OVER (PARTITION BY cd.coin_id ORDER BY cd.date) AS coins_percentage_change,
                   ((s.close - LAG(s.close, 1) OVER (PARTITION BY s.ticker ORDER BY s.date)) * 100.0) / LAG(s.close, 1) OVER (PARTITION BY s.ticker ORDER BY s.date) AS IXIC_percentage_change
                   FROM
                   coins_data AS cd
                   INNER JOIN
                   stocks AS s ON cd.date = s.date
                   WHERE
                   cd.coin_id = 'ethereum' AND s.ticker = 'IXIC'
                   ORDER BY cd.date DESC;""", con=connection)
         st.dataframe(df)
     if option =="Compare top 3 coins daily price trend vs Nifty (^NSEI)":
         df=pd.read_sql("""WITH Bitcoin_Change AS (
         SELECT
         cd.date,
         ((cd.price_usd - LAG(cd.price_usd, 1) OVER (PARTITION BY cd.coin_id ORDER BY cd.date)) * 100.0) / LAG(cd.price_usd, 1) OVER (PARTITION BY cd.coin_id ORDER BY cd.date) AS Bitcoin_percentage_change
         FROM
         coins_data AS cd
         WHERE
         cd.coin_id = 'bitcoin'
         ),
         Ethereum_Change AS (
         SELECT
         cd.date,
         ((cd.price_usd - LAG(cd.price_usd, 1) OVER (PARTITION BY cd.coin_id ORDER BY cd.date)) * 100.0) / LAG(cd.price_usd, 1) OVER (PARTITION BY cd.coin_id ORDER BY cd.date) AS Ethereum_percentage_change
         FROM
         coins_data AS cd
         WHERE
         cd.coin_id = 'ethereum'
         ),
         Tether_Change AS (
         SELECT
         cd.date,
         ((cd.price_usd - LAG(cd.price_usd, 1) OVER (PARTITION BY cd.coin_id ORDER BY cd.date)) * 100.0) / LAG(cd.price_usd, 1) OVER (PARTITION BY cd.coin_id ORDER BY cd.date) AS Tether_percentage_change
         FROM
         coins_data AS cd
         WHERE
         cd.coin_id = 'tether'
         ),
         NSEI_Change AS (
         SELECT
         s.date,
         ((s.close - LAG(s.close, 1) OVER (PARTITION BY s.ticker ORDER BY s.date)) * 100.0) / LAG(s.close, 1) OVER (PARTITION BY s.ticker ORDER BY s.date) AS NSEI_percentage_change
         FROM
         stocks AS s
         WHERE
         s.ticker = '^NSEI'
         )
         SELECT
         COALESCE(bc.date, ec.date, tc.date, nc.date) AS date,
         bc.Bitcoin_percentage_change,
         ec.Ethereum_percentage_change,
         tc.Tether_percentage_change,
         nc.NSEI_percentage_change
         FROM
         Bitcoin_Change AS bc
         LEFT JOIN
         Ethereum_Change AS ec ON bc.date = ec.date
         LEFT JOIN
         Tether_Change AS tc ON COALESCE(bc.date, ec.date) = tc.date
         LEFT JOIN
         NSEI_Change AS nc ON COALESCE(bc.date, ec.date, tc.date) = nc.date
         ORDER BY
         date """, con=connection)
         df = df.round(2)
         st.dataframe(df)
     if option =="Compare stock prices (^GSPC) with crude oil prices on the same dates":
         df = pd.read_sql( """
         SELECT
         s.date,
         s.close AS sp500_price,
         ((s.close - LAG(s.close,1) OVER (PARTITION BY s.ticker ORDER BY s.date)) * 100.0) / LAG(s.close,1) OVER (PARTITION BY s.ticker ORDER BY s.date) AS sp500_percentage_change,
         od.price_usd AS oil_price,
         ((od.price_usd - LAG(od.price_usd,1) OVER (ORDER BY od.date)) * 100.0) / LAG(od.price_usd,1) OVER (ORDER BY od.date) AS oil_percentage_change
         FROM
         stocks AS s
         INNER JOIN
         oil_datas AS od ON s.date = od.date
         WHERE
         s.ticker = '^GSPC'
         ORDER BY
         s.date DESC;
         """, con = connection)
         df = df.round(2)
         st.dataframe(df)
     if option =="Compare NASDAQ (^IXIC) with Ethereum price trends":
         df = pd.read_sql( """
         SELECT
         s.date,
         s.close AS NSDAQ_price,
         ((s.close - LAG(s.close,1) OVER (PARTITION BY s.ticker ORDER BY s.date)) * 100.0) / LAG(s.close,1) OVER (PARTITION BY s.ticker ORDER BY s.date) AS NSDAQ_percentage_change,
         co.price_usd AS Ethereum_price,
         ((co.price_usd - LAG(co.price_usd,1) OVER (PARTITION BY co.coin_id ORDER BY co.date)) * 100.0) / LAG(co.price_usd,1) OVER (PARTITION BY co.coin_id ORDER BY co.date) AS ETH_percentage_change
         FROM
         stocks AS s
         INNER JOIN
         coins_data AS co ON s.date = co.date
         WHERE
         s.ticker = 'IXIC' AND co.coin_id = 'ethereum'
         ORDER BY
         s.date;
         """, con = connection)
         df = df.round(2)
         st.dataframe(df)
     if option =="Correlate Bitcoin closing price with crude oil closing price (same date)":
         df_coins= pd.read_sql("SELECT date,price_usd AS BTC_price from coins_data WHERE coin_id= 'bitcoin'", con=connection)
         df_oil= pd.read_sql("SELECT date,price_usd AS oil_price from oil_datas", con=connection)
         df = pd.merge(df_coins, df_oil, on="date")
         df['date']=pd.to_datetime(df['date'])
         percentage_df = df[['BTC_price', 'oil_price']].pct_change() * 100
         percentage_df.columns = ['BTC_price_pct_change', 'oil_price_pct_change']
         rolling_correlation_series = percentage_df['BTC_price_pct_change'].rolling(window=7).corr(percentage_df['oil_price_pct_change'])
         rolling_correlation_df = pd.DataFrame({
         'date': df['date'],
         'rolling_correlation_7_day': rolling_correlation_series
          })

         compare1 = pd.merge(df, percentage_df.drop(columns='date', errors='ignore'), left_index=True, right_index=True, how='left')
         compare = pd.merge(compare1, rolling_correlation_df, on='date', how='left')

         df1 =pd.DataFrame(compare)
         st.dataframe(df1)
     if option =="Top 3 crypto coins with stock indices for 2025":
        df = pd.read_sql("""
              SELECT
              cp.date,
              cp.coin_id,
              cp.price_usd,
              sp.ticker,
              sp.close
              FROM
              coins_data AS cp
              join
              stocks AS sp ON cp.date = sp.date
              WHERE cp.date BETWEEN '2025-01-01' AND '2026-01-01'

              """, con = connection)
        st.dataframe(df)

     if option =="stock prices, oil prices, and Bitcoin prices for daily comparison":
         df = pd.read_sql( """
                      SELECT
                      cp.date,
                      cp.price_usd AS btc_price,
                      op.price_usd AS oil_price,
                      sp.close AS sp500
                      FROM
                      coins_data AS cp
                      join
                      oil_datas AS op ON cp.date = op.date
                      join
                      stocks AS sp ON cp.date = sp.date
                      WHERE cp.coin_id = 'bitcoin' AND sp.ticker = '^GSPC'

         """, con = connection)

         st.dataframe(df)

     if option == "Display date":
         df = pd.read_sql("SELECT coins_data.price_usd AS BTC_USD, stocks.close AS close_price FROM coins_data INNER JOIN stocks ON coins_data.date = stocks.date WHERE coins_data.coin_id = 'bitcoin' AND stocks.ticker = '^GSPC'",con=connection)
         corr= df['BTC_USD'].corr(df['close_price'])
         st.metric(label = "correlation", value ="corr" )

if selected == "Top 3 crypto Analysis":
    st.title("Top 3 Crypto Analysis")
    st.write("Daily Price Analysis for Top crypto currencies ")

    col1, col2 = st.columns(2)

    with col1:
        start_date = st.date_input("Start Date", value=default_start_crypto)

    with col2:
        end_date = st.date_input("End Date", value=default_end_crypto)

    if start_date > end_date:
        st.error("Error: End date must be after start date.")
    else:
        if selected == "Top 3 crypto Analysis":
           option = st.selectbox(
           "Select a query",(
           "bitcoin",
           "Ethereum",
           "Tether")
           )
        if option == "bitcoin":
           df = pd.read_sql(f"SELECT date,price_usd FROM coins_data WHERE coin_id = 'bitcoin' AND date BETWEEN '{start_date}' AND '{end_date}'", con=connection)
           st.dataframe(df)
        if option == "Tether":
           df = pd.read_sql(f"SELECT date,price_usd FROM coins_data WHERE coin_id = 'tether' AND date BETWEEN '{start_date}' AND '{end_date}'", con=connection)
           st.dataframe(df)
        if option == "Ethereum":
           df = pd.read_sql(f"SELECT date,price_usd FROM coins_data WHERE coin_id = 'ethereum' AND date BETWEEN '{start_date}' AND '{end_date}'", con=connection)
           st.dataframe(df)
```
