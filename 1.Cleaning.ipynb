**Z-Score method to find outlier**

```python
import numpy as np
from scipy.stats import zscore

data_Hillshade = data["Hillshade_9am"]
z_scores = zscore(data_Hillshade)
print(z_scores)
outliers = np.where(np.abs(z_scores) > 3)
print(outliers)
print(f"length of outlier: {len(outliers[0])}")
```

### Quantifying Outliers using IQR Method

```python
for col in numerical_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Corrected usage of np.where to cap lower outliers
    data[col] = np.where(data[col] < lower_bound, lower_bound, data[col])
    # Corrected usage of np.where to cap upper outliers
    data[col] = np.where(data[col] > upper_bound, upper_bound, data[col])
```

```python
numerical_cols = ['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology',
       'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways',
       'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
       'Horizontal_Distance_To_Fire_Points']
# Loop through numerical columns and create boxplots

plt.figure(figsize=(15, 7))
plt.suptitle("Boxplots before Outliers removal")
for i in range(0, len(numerical_cols)):
    plt.subplot(2, 5, i+1)
    sns.boxplot(y=data[numerical_cols[i]],color='purple')
    plt.tight_layout()
```
